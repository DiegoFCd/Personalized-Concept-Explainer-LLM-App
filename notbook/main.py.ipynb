## Crear el Prompt Template

```python
prompt_template_str = """
Tu tarea es explicarme el concepto de **{concept}** de una manera que sea:

1. Clara e intuitiva
2. Concisa (en menos de 100 palabras)
3. Adaptada específicamente a mí y a lo que ya sé

Usa la siguiente información sobre mí para personalizar tu explicación:

- Rol: BI & Data Analyst en transición hacia Data Scientist
- Conocimientos: Machine Learning (modelos supervisados, clustering, clasificación), ETL y visualización de datos.
- Objetivo: Entender conceptos de LLMs como extensión de mi base en ML.

La personalización debe ser sutil y natural.
Evita referencias forzadas a mi background.
"""
```

## Asocio la variable creada al str

```python
prompt_template = PromptTemplate.from_template(
    prompt_template_str
)
```

## Crear la función

```python
def generate_explanation(concept):

    prompt = prompt_template.format(
        concept=concept
    )

    response = model.invoke(prompt)

    return response.content
```

## Probar la función

```python
generate_explanation("aprendizaje no supervisado")
```

## Crear la interfaz Gradio

```python
demo = gr.Interface(
    fn= generate_explanation,
    inputs=[
        gr.Textbox(
            label = "Concepto",
            lines = 1)
    ],
    outputs=[
        gr.Textbox(
        label = "Explicacion",
        lines = 5)
    ],
    flagging_mode = "never",
    title = "Definiciones Personalizadas",
    description = "Ingresa un concepto y obtén una explicación adaptada a tu perfil."
)
```

## Lanzar la aplicación

```python
demo.launch()
```
